```python
# 데이터 로드 및 정보 / 요약통계량 확인
df = pd.read_csv(path + r'\거시경제지표.csv', encoding = 'cp949')
df.info()
df.describe()
```

## **PART4. 그래프 생성**

```python
# Seaborn 스타일 설정
sns.set_style('whitegrid')

# 한글폰트 관련 패치
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

#### **PART4-1. 일반적인 산점도 행렬**

```python
# 조회할 변수 list 생성
vars = ['부실순증금액', '선행종합지수', 'USD리보금리', '장단기금리차(미국)', '근원물가지수', '원달러환율']

# grid 객체 생성
# sns.pairplot은 여타 그래프 객체와 다르게, 도화지 / 좌표계 / 그래프 객체 일체형임
grid = sns.pairplot(data = df, dropna = True, vars = vars,
                    kind = 'reg', diag_kind = 'kde', height = 2)
```

#### **PART4-2. 코로나19 전후 구분(hue= 옵션 적용)**

```python
grid = sns.pairplot(data = df, dropna = True, vars = vars, hue = '코로나구분',
                    kind = 'reg', diag_kind = 'kde',height = 2, palette = 'Set2')

# .map_lower() / .map_diag() / .map_upper() 메서드를 통해 격자의 하방 / 대각 / 상방 그래프 별도 생성 가능
grid.map_lower(sns.kdeplot);
```
