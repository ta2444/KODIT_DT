### **(참고1) 리스트 컴프리헨션 연습**
#### **정의**
* 반복문(for문)과 제어문(if / if-else)을 이용해 리스트를 효율적으로 생성하는 기법
    * 참고로, if와 if-else는 아예 다른 문장임
        * if : 필터의 역할 => 참일 때 값만 출력함
        * A if (조건문) else B : 분기의 역할 => 조건문이 참이면 A를 출력, 거짓이면 B를 출력
 
#### **문법**
#### **(반복문만 쓸 경우)**
[ i   for i in range(시작, 끝)]
#### **(if문도 쓸 경우)**
[ i   for i in range(시작, 끝) **if** 조건문]
#### **(if-else 삼항연산 쓸 경우)**
[참일 때 취할 값 **if** 조건문 **else** 거짓일 때 취할 값 for i in range(시작, 끝)]

```python
# 0부터 11까지 출력
[i for i in range(0, 12)]
```

```python
# 0부터 11까지 출력하되, /2 적용
[i/2 for i in range(0, 12)]
```

```python
# 홀수만 출력해보기
[i for i in range(0, 12) if i%2 == 0]
```

```python
# 홀수면 100을 곱하고, 짝수면 10을 곱해서 출력
[i * 100 if i%2 == 0 else i * 10 for i in range(0, 12)]
```

### **(참고2) f-string 연습**
#### **문법**
#### f'접두사 {변수명 : 표시형식} 접미사'}

```python
x = 1.5555555555
print(f' 정수 : {x : <0.0f}')
print(f' 소숫점 첫째자리 : {x : <0.1f}')
print(f' 소숫점 둘째자리 : {x : <0.2f}')
print(f' 소숫점 셋째자리 : {x : <0.3f}')
print(f' 소숫점 넷째자리 : {x : <0.4f}')
print(f' 지수표현식 : {x : <0.0}')
```

```python
year = [i for i in range(2020, 2024)]
month = [i for i in range(1, 13)]

for yy in year:
    for mm in month:
        print(f'{yy}년{mm : 1}월')
```

### **(참고3) pd.Series.isin() 메서드**
* 해당 시리즈(DataFrame의 특정 열)의 값이 특정 집합에 포함되었는지 여부를 반환하는 함수
* ex) df['계정구분'].isin(['소위탁'])
* 👉df['계정구분']의 자료형은 Series이고, 해당 Series에 포함된 요소(관측치)의 값이 '소위탁'에 해당하면 **True**, 아니면 **False**를 반환함
* </br>
* '~'을 조건문 맨 앞에 붙이면, True와 False의 값이 반전됨 (Not의 의미)
* ex) ~df['계정구분'].isin(['소위탁'])
* * 👉df['계정구분']의 자료형은 Series이고, 해당 Series에 포함된 요소(관측치)의 값이 '소위탁'에 해당하면 **False**, 아니면 **True** 반환함

```python
df['계정구분'].isin(['소위탁'])
```

```python
~df['계정구분'].isin(['소위탁'])
```

## **PART4. CHALLENGING**🔥 - 계정별 분리해서 그래프 생성(subplot 활용)

```python
fig = plt.figure(figsize = [12, 5])

ax1 = plt.subplot(211)
plt.tight_layout(h_pad = 5, w_pad = 0)
plot1 = sns.lineplot(data = df[df['계정구분'] == '일반'], x = '연도', y = '부실률'
                     , linewidth = 3, color=  'r')
plt.title("일반보증 부실률 현황", fontweight = 'bold', fontsize = 15)
plt.xlabel(' ')
plt.ylabel('')
plt.xticks(rotation = -30)
plt.yticks(ticks = [i / 2 for i in range(2, 10)], labels = [f'{i / 2 :.1f}' for i in range(2, 10)])

ax2 = plt.subplot(223)
plot2 = sns.lineplot(data = df[df['계정구분'] == '소위탁'], x = '연도', y = '부실률'
                     , linewidth = 3, color=  'g')
plt.title("소위탁 부실률 현황", fontweight = 'bold', fontsize = 15)
plt.xticks(rotation = -30)
plt.yticks(ticks = [i for i in range(0, 13, 2)], labels = [f'{i :.1f}' for i in range(0, 13, 2)])
plt.xlabel('')
plt.ylabel('')

ax3 = plt.subplot(224)
plt.title("대환보증 부실률 현황", fontweight = 'bold', fontsize = 15)
plot2 = sns.lineplot(data = df[df['계정구분'] == '대환'], x = '연도', y = '부실률'
                     , linewidth = 3, color=  'b')
plt.xticks(rotation = -30)
plt.yticks(ticks = [i for i in range(0, 6, 1)], labels = [f'{i :.1f}' for i in range(0, 6, 1)])
plt.xlabel('')
plt.ylabel('');
```
