# **04. boxplot(상자그림) 실습**

```python
# 필요 패키지 호출
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import re

# 실습용 데이터가 저장된 폴더
path = r'C:\HANSSAK\SecureGate\Download\DT_Python 시각화'
print('작업 폴더 주소 : ', path)

# 출력 옵션 지정
pd.set_option('display.width', 2000)                  # 한줄에 표시할 컬럼 관련
pd.set_option('display.max_colwidth', None)           # 최대 컬럼너비 (None 지정시 데이터 생략 없음)
pd.options.display.float_format = '{:.2f}'.format     # 표시할 숫자 형식 (지수형태 표현에서 소숫점 2자리까지로)
```

```python
# 기업체명 비식별화 함수
def remove(x):
    if x.find("(") == 0:
        y = x[0: x.index(")") + 3].strip()
    else:
        y = x[0:2]

    return y
```

```python
#데이터 불러오기
df = pd.read_csv(path + r'\sample_classifying.csv', encoding= 'cp949')
                 
df.fillna("", inplace = True)
df.info()
df.head(10)
```

```python
# 함수1 : 업체명 맨 앞글자 '(주)','(유한)' 등 기업형태 문자열 제거
def clear(x):
    if x.find("(") == 0:
        y = x[x.index(")") + 1 : len(x)]
    else:
        y = x
        
    return y

# 함수2 : 정규표현식 활용, 업체명 맨 앞글자를 가나다.../숫자/영문/특수문자로 분류 => 지점 편의에 따라 수정
def ganada(x):
    gubun = ['^[가-낗]', '^[나-닣]', '^[다-띻]', '^[라-맇]', '^[마-밓]', '^[바-삫]', '^[사-앃]', '^[아-잏]',
             '^[자-찧]', '^[차-칳]', '^[카-킿]', '^[타-팋]', '^[파-핗]', '^[하-힣]', '^[A-z]',  '^[0-9]', 
             '^[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]']

    for i, char in enumerate(gubun):
        m = re.search(char, x)
        if m != None:
            if i < 14:
                y = char[2]
                # print(i)
            elif i == len(gubun)-3:
                y = '힣_영어'
            elif i == len(gubun)-2:
                y = '힣_숫자'
            else:
                y = '힣_특수'
            return(y)
```

```python
# 데이터에 함수1, 2 적용
df['기업체명'] = df['기업체명'].apply(clear)
df['구분'] = df['기업체명'].apply(ganada)

# 업체수 계산용 - 중복제거한 데이터셋 생성
df_up = df[['기업체명', '구분']].drop_duplicates().reset_index(drop = True)

df.head(10)
```

```python
# 원장수, 업체수, 보증잔액 집계 데이터프레임 생성
df_pivot = pd.pivot_table(data = df, index = '구분', aggfunc={'원장번호':'count', '보증잔액':'sum'})
df_up_pivot = pd.pivot_table(data = df_up, index = '구분', aggfunc={'기업체명':'count'})

# 하나로 합치기(OUTER FULL JOIN)
df_pivot = pd.merge(left = df_pivot, right = df_up_pivot, how = 'outer', on = '구분').reset_index(drop = False)
df_pivot.columns = ['구분', '보증잔액_합계', '원장수', '업체수']
## '구분'컬럼의 순서를 위해 임시로 넣었던 '힣_' 문자열 제거
df_pivot['구분'] = df_pivot['구분'].apply(lambda x : x.replace('힣_', '')) 
df_pivot
```

```python
# 총 업체수/원장수/보증잔액 계산 (누적 비중 계산용)
total_up = df_pivot['업체수'].sum()
total_cnt = df_pivot['원장수'].sum()
total_jan = df_pivot['보증잔액_합계'].sum()

# 누적 비중 계산
업체수_누적_비중 = []
원장수_누적_비중 = []
보증잔액_누적_비중 = []

up_ratio = 0
cnt_ratio = 0
jan_ratio = 0
변수 = ['']

# 반복문을 통해 누적 비중 계산
for i in range(len(df_pivot)):
    up_ratio+=(df_pivot['업체수'][i] / total_up) * 100
    cnt_ratio+=(df_pivot['원장수'][i] / total_cnt) * 100
    jan_ratio+=(df_pivot['보증잔액_합계'][i] / total_jan) * 100
    
    업체수_누적_비중.append(up_ratio)
    원장수_누적_비중.append(cnt_ratio)
    보증잔액_누적_비중.append(jan_ratio)

# 하나로 합치기 => 최종 데이터셋 df_total 생성
df_add = pd.DataFrame(zip(업체수_누적_비중, 원장수_누적_비중, 보증잔액_누적_비중), columns = ['업체수_누적_비중', '원장수_누적_비중', '보증잔액_누적_비중'])

df_total = pd.concat([df_pivot, df_add], axis = 1)

df_total
```

```python
# 파이 차트(Pie Chart) 그리기
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False

fig, ax = plt.subplots(figsize=(8, 8))

# wedgeprops={'width': 0.5, 'edgecolor': 'w', 'linewidth': 2}

theme = plt.get_cmap('pink')

ax.set_prop_cycle("color", [theme((1- 1. * i / len(df_total['구분']))-0.3) for i in range(len(df_total['구분']))])
ax.pie('업체수', 
       labels='구분',
#        autopct = '%.1f%%',
       labeldistance=1.1, data = df_total,
        counterclock= False,
        textprops=dict(color="w", fontsize = 13),
        startangle=-270,
        );

labels = [f'{l}, {s:0.1f}%' for l, s in zip(df_total['구분'], 100 * df_total['업체수'] / total_up)]
ax.legend(labels,
          bbox_to_anchor=(1, 0, 1, 1),
          title = '업체수 기준',
          loc="center left",
          fontsize = "13"
          );
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (20,5), sharey = True)
fig.tight_layout()
axs = [ax1, ax2, ax3]
labels = ['업체수', '원장수', '보증잔액']

for i, a in enumerate(axs):
    # 제목 지정
    a.set_title(f'{labels[i]} 기준', fontdict = {'size' : 14})
    # 33% 선
    a.axhline(y = 33, color = 'r')
    a.text(0, 35, '33%', fontsize = 10)
    # 66% 선
    a.axhline(y = 66, color = 'r')
    a.text(0, 68, '66%', fontsize = 10)
    a.set_xticklabels(df_total['구분'], fontsize=14)
    # 막대그래프
    a.bar(height = f'{labels[i]}_누적_비중', x = '구분', data = df_total);
```
