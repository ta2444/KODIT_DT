## **PART3. Îç∞Ïù¥ÌÑ∞ ÏßëÍ≥Ñ**

* Ï£ºÏÇ¨ÏóÖÏû•ÏÜåÏû¨ÏßÄ(ÏãúÎèÑ)Î≥ÑÎ°ú Î∂ÄÏ±ÑÎπÑÏú®, ÏòÅÏóÖÏù¥ÏùµÎ•†Ïùò Ï§ëÏúÑÏàòÏôÄ ÏóÖÏ≤¥ÏàòÎ•º ÏßëÍ≥Ñ
    * Ïû¨Î¨¥ÎπÑÏú®Ïùò ÏßëÍ≥ÑÌï®ÏàòÎäî Ï§ëÏúÑÏàò(median) ÏÇ¨Ïö©
    * ÏóÖÏ≤¥ÏàòÎäî Í≥†Í∞ùÎ≤àÌò∏Ïùò Í∞úÏàòÎ•º ÏÑ∏Îäî Î∞©Î≤ï (count Ìï®Ïàò) ÏÇ¨Ïö©
      
* pd.pivot_table()Î©îÏÑúÎìúÏùò aggfunc = ÌååÎùºÎØ∏ÌÑ∞Ïùò Í∞íÏúºÎ°ú ÎîïÏÖîÎÑàÎ¶¨Î•º Ï†ÑÎã¨ÌïòÎ©¥ Ïª¨ÎüºÎ≥ÑÎ°ú Îã§Î•∏ Ìï®Ïàò Ï†ÅÏö© Í∞ÄÎä•
    * ex) aggfunc = {'Ïª¨ÎüºÎ™Ö1' : 'ÏßëÍ≥ÑÌï®Ïàò1', 'Ïª¨ÎüºÎ™Ö2' : 'ÏßëÍ≥ÑÌï®Ïàò2'}
    * Ï∞∏Í≥†) Ï£ºÏöî ÏßëÍ≥ÑÌï®Ïàò Ï¢ÖÎ•ò:
        * sum (Ìï©Í≥Ñ)
        * count (Í∞úÏàò)
        * mean (ÌèâÍ∑†)
        * median (Ï§ëÏúÑÏàò)
        * mode (ÏµúÎπàÍ∞í)
        * std (ÌëúÎ≥∏ÌëúÏ§ÄÌé∏Ï∞®)
        * var (ÌëúÎ≥∏Î∂ÑÏÇ∞)
        * min (ÏµúÏÜüÍ∞í)
        * max (ÏµúÎåìÍ∞í)

```python
df_agg = pd.pivot_table(data = df, index = ['ÏßÄÏó≠'], 
                        values=['Î∂ÄÏ±ÑÎπÑÏú®', 'ÏòÅÏóÖÏù¥ÏùµÎ•†', 'Ï¥ùÏûêÏÇ∞ÌöåÏ†ÑÏú®', 'Í≥†Í∞ùÎ≤àÌò∏'], 
                        aggfunc = {'Î∂ÄÏ±ÑÎπÑÏú®' : 'median', 
                                   'ÏòÅÏóÖÏù¥ÏùµÎ•†' : 'median', 
                                   'Ï¥ùÏûêÏÇ∞ÌöåÏ†ÑÏú®' : 'median', 
                                   'Í≥†Í∞ùÎ≤àÌò∏' : 'count'}).reset_index(drop= False)

# Ïª¨ÎüºÎ™Ö Î≥ÄÍ≤Ω
df_agg.rename({'Í≥†Í∞ùÎ≤àÌò∏' : 'ÏóÖÏ≤¥Ïàò'}, axis = 1, inplace = True)

# ÏóÖÏ≤¥Ïàò ÎßéÏùÄ ÏàúÏÑúÎåÄÎ°ú Ï†ïÎ†¨
df_agg.sort_values(by = 'ÏóÖÏ≤¥Ïàò', ascending= False, inplace = True)

# ÏóÖÏ≤¥Ïàò ÌïòÏúÑ 3Í∞ú ÏßÄÏó≠ Ï†úÍ±∞ 
df_agg = df_agg[~df_agg['ÏßÄÏó≠'].isin(['ÎØ∏ÏÉÅ', 'ÏÑ∏Ï¢Ö', 'Ï†úÏ£º'])]

df_agg
```

### **(Ï∞∏Í≥†) pd.Series.isin() Î©îÏÑúÎìú**
* Ìï¥Îãπ ÏãúÎ¶¨Ï¶à(DataFrameÏùò ÌäπÏ†ï Ïó¥)Ïùò Í∞íÏù¥ ÌäπÏ†ï ÏßëÌï©Ïóê Ìè¨Ìï®ÎêòÏóàÎäîÏßÄ Ïó¨Î∂ÄÎ•º Î∞òÌôòÌïòÎäî Ìï®Ïàò
* ex) df['ÏßÄÏó≠'].isin(['ÏÑúÏö∏'])
* üëâdf['ÏßÄÏó≠']Ïùò ÏûêÎ£åÌòïÏùÄ SeriesÏù¥Í≥†, Ìï¥Îãπ SeriesÏóê Ìè¨Ìï®Îêú ÏöîÏÜå(Í¥ÄÏ∏°Ïπò)Ïùò Í∞íÏù¥ 'ÏÑúÏö∏'Ïóê Ìï¥ÎãπÌïòÎ©¥ True, ÏïÑÎãàÎ©¥ FalseÎ•º Î∞òÌôòÌï®
* '~'ÏùÑ Ï°∞Í±¥Î¨∏ Îß® ÏïûÏóê Î∂ôÏù¥Î©¥, TrueÏôÄ FalseÏùò Í∞íÏù¥ Î∞òÏ†ÑÎê® (NotÏùò ÏùòÎØ∏)

```python
df['ÏßÄÏó≠'].isin(['ÏÑúÏö∏'])
```

```python
~df['ÏßÄÏó≠'].isin(['ÏÑúÏö∏'])
```

### **(Ï∞∏Í≥†) Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏÑúÎ∏åÏÑ∏ÌåÖ(Subsetting)**
