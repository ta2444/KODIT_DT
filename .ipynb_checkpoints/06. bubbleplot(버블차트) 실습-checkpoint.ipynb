## **PART3. 데이터 집계**

* 주사업장소재지(시도)별로 부채비율, 영업이익률의 중위수와 업체수를 집계
    * 재무비율의 집계함수는 중위수(median) 사용
    * 업체수는 고객번호의 개수를 세는 방법 (count 함수) 사용
      
* pd.pivot_table()메서드의 aggfunc = 파라미터의 값으로 딕셔너리를 전달하면 컬럼별로 다른 함수 적용 가능
    * ex) aggfunc = {'컬럼명1' : '집계함수1', '컬럼명2' : '집계함수2'}
    * 참고) 주요 집계함수 종류:
        * sum (합계)
        * count (개수)
        * mean (평균)
        * median (중위수)
        * mode (최빈값)
        * std (표본표준편차)
        * var (표본분산)
        * min (최솟값)
        * max (최댓값)

```python
df_agg = pd.pivot_table(data = df, index = ['지역'], 
                        values=['부채비율', '영업이익률', '총자산회전율', '고객번호'], 
                        aggfunc = {'부채비율' : 'median', 
                                   '영업이익률' : 'median', 
                                   '총자산회전율' : 'median', 
                                   '고객번호' : 'count'}).reset_index(drop= False)

# 컬럼명 변경
df_agg.rename({'고객번호' : '업체수'}, axis = 1, inplace = True)

# 업체수 많은 순서대로 정렬
df_agg.sort_values(by = '업체수', ascending= False, inplace = True)

# 업체수 하위 3개 지역 제거 
df_agg = df_agg[~df_agg['지역'].isin(['미상', '세종', '제주'])]

df_agg
```

### **(참고) pd.Series.isin() 메서드**
* 해당 시리즈(DataFrame의 특정 열)의 값이 특정 집합에 포함되었는지 여부를 반환하는 함수
* ex) df['지역'].isin(['서울'])
* 👉df['지역']의 자료형은 Series이고, 해당 Series에 포함된 요소(관측치)의 값이 '서울'에 해당하면 True, 아니면 False를 반환함
* '~'을 조건문 맨 앞에 붙이면, True와 False의 값이 반전됨 (Not의 의미)

```python
df['지역'].isin(['서울'])
```

```python
~df['지역'].isin(['서울'])
```

### **(참고) 데이터프레임 서브세팅(Subsetting)**
* 전체 데이터프레임 중 특정 조건을 만족하는 행이나 필요한 열만 남기는 것

#### (문법) 데이터프레임명[행을 거를 조건문][남길열(list)]

```python
df[df['업종대분류코드'] == 'C'][['지역', '영업이익률', '총자산회전율']]
```

```python
df[df['업종대분류코드'].isin(['C', 'F'])][['지역', '영업이익률']]
```

## **PART4. 그래프 생성**

```python
# Seaborn 스타일 설정
sns.set_style('whitegrid')

# 한글폰트 관련 패치
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
```

#### **PART4-1. 일반적인 산점도 행렬**

```python
# 도화지 / 좌표계 / 그래프 객체 생성
fig = plt.figure(figsize = [7, 5])
ax = sns.scatterplot(data = df_agg, x = '부채비율', y = '총자산회전율', size = '업체수', hue = '지역',
                     sizes = (100, 5000), legend = 'auto', palette = 'bright')
plt.xticks([i for i in range(100, 200, 10)])
plt.ylim(3, 6)

handles, labels  =  ax.get_legend_handles_labels()
plt.legend(handles[:-5], labels[:-5], loc='lower right', title = '지역별', bbox_to_anchor = (1.2, 0), ncol =1 )
```
